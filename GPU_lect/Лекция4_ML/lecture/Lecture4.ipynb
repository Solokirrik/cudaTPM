### Подключаем необходимые библиотеки

```python
from sklearn import linear_model
```

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
```

* http://matplotlib.org/gallery.html

### Загружаем данные из CSV файла

```python
Location = './Data/MeanT.csv'
df = pd.read_csv(Location)
```

```python
df.head()
```

### Подготавливаем данные

```python
X = df['Year'].values
X = X.reshape((X.shape[0],1))

y = df['Temperature'].values
```

### Обучаем модель

```python
reg = linear_model.LinearRegression()
reg.fit(X, y) 
```

### Сравниваем полученный результат с исходными данными

```python
years = df['Year'].values
```

```python
result = reg.predict(years.reshape((years.shape[0],1)))
```

```python
plt.title('Result analysis')
plt.plot(years,result)
plt.plot(years,y)
plt.xlabel('Years')
plt.ylabel('Temperature')
plt.show()
```

### Что же нас ждёт?

```python
next_years = np.linspace(1880,2150)
next_result = reg.predict(next_years.reshape((next_years.shape[0],1)))
```

```python
plt.title('Next years')
plt.plot(next_years,next_result)
plt.plot(years,y)
plt.xlabel('Years')
plt.ylabel('Temperature')
plt.show()
```
