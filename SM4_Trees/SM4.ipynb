```python
def get_min_Gini(X1, crit_Gini, X_len):
    for i in range(0, X_len):
        beta = X1[1][i]

        mask_y_0 = X1[0] <= 0
        mask_y_1 = X1[0] > 0

        mask_x_r = X1[1] > beta
        mask_x_l = X1[1] <= beta

        len_l = mask_x_l.sum()
        len_r = X_len - len_l
        if not ((len_l) and (len_r)):
            crit_Gini[0][i] = float("inf")
        else:
            p0l = (mask_x_l * mask_y_0).sum() / len_l
            p0r = (mask_x_r * mask_y_0).sum() / len_r
            p1l = (mask_x_l * mask_y_1).sum() / len_l
            p1r = (mask_x_r * mask_y_1).sum() / len_r

            pxl = mask_x_l.sum() / len_l
            pxr = mask_x_r.sum() / len_r

            l_H = 2 * p0l * p1l
            r_H = 2 * p0r * p1r

            crit_Gini[0][i] = (pxl / X_len) * l_H + (pxr / X_len) * r_H
        if i % 1 == 0:
            print(i, crit_Gini[0][i])


t1 = time.time()
get_min_Gini(X1, crit_Gini, X_len)
print(crit_Gini.min(), crit_Gini.argmin())
print(time.time() - t1)
```

```python
# coding=utf-8
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold


df = pd.read_csv('./data/cs-training.csv', sep=',')

df = df.dropna()
X0 = df.as_matrix(columns=df.columns[:])
X = df.as_matrix(columns=df.columns[1:])
y = df.as_matrix(columns=df.columns[:1])
y = y.reshape(y.shape[0])

gkf = KFold(n_splits=5, shuffle=True)
# print(X.shape, y.shape)
# print(X.shape, y.shape)

def init_list_of_arrs(size):
    list_of_arrs = list()
    for i in range(0, size):
        list_of_arrs.append(np.zeros((1, 1)))
    return list_of_arrs

def get_b_ns(y_vec, y_beta, inp_vec, perc=0.03):
    w_vec = inp_vec
    if w_vec.shape[0] > 1000:
        btch_sz = int(w_vec.shape[0] * perc)
    else:
        btch_sz = w_vec.shape[0]
    arr_of_numbs = np.linspace(0, w_vec.shape[0] - 1, btch_sz, dtype=int)
    item_beta_g_arr = np.zeros((btch_sz, 3))
    iter_c = 0
    for item in arr_of_numbs:
        beta = w_vec[item]

        xl_len = (w_vec[arr_of_numbs] <= beta).sum()
        xr_len = btch_sz - xl_len
        if xl_len != 0 and xr_len != 0:

            p0l = ((y_vec[arr_of_numbs] <= y_beta) * (w_vec[arr_of_numbs] <= beta)).sum() / xl_len
            p0r = ((y_vec[arr_of_numbs] <= y_beta)*(w_vec[arr_of_numbs] > beta)).sum() / xr_len
            p1l = ((y_vec[arr_of_numbs] > y_beta) * (w_vec[arr_of_numbs] <= beta)).sum() / xl_len
            p1r = ((y_vec[arr_of_numbs] > y_beta)*(w_vec[arr_of_numbs] > beta)).sum() / xr_len

            # hl = p0l * p1l
            # hr = p0r * p1r
            if p0l == 0:
                hl = -p1l * np.log(p1l)
            elif p1l == 0:
                hl = -p0l * np.log(p0l)
            else:
                hl = -p0l * np.log(p0l) - p1l * np.log(p1l)
            if p0r == 0:
                hr = -p1r * np.log(p1r)
            elif p1r == 0:
                hr = -p0r * np.log(p0r)
            else:
                hr = -p0r * np.log(p0r) - p1r * np.log(p1r)

            item_beta_g_arr[iter_c][0] = item
            item_beta_g_arr[iter_c][1] = beta
            G_l = (xl_len / btch_sz) * hl
            G_r = (xr_len / btch_sz) * hr
            G = G_l + G_r
            item_beta_g_arr[iter_c][2] = G
        else:
            item_beta_g_arr[iter_c][2] = 10
        iter_c += 1
    bestG = item_beta_g_arr.T[2].argmin()
    out = np.array([item_beta_g_arr[bestG][1], item_beta_g_arr[bestG][2]])
    # return item_beta_g_arr[item_beta_g_arr.T[2].argmin()]
    return out


# list_size = pow(2, X.shape[1] + 1) - 1
list_size = 10000
# list_size = 2047
arrList = init_list_of_arrs(list_size)
best_features_beta = np.zeros((list_size, 2))

features = X.shape[1]
features_list = [i for i in range(features)]
b_g_arr = np.zeros((features, 2))

arrList[0] = X
print(0, arrList[0].shape)
print("-----------")
iter = 0
for i in range(4095):
    if arrList[i].shape[0] != 0:
        if arrList[i].shape[0] == 1:
            pass
        else:
            for feature in features_list:
                b_g_arr[feature] = get_b_ns(y, 0, arrList[i].T[feature])
            bst_feat = b_g_arr.T[1].argmin()
            beta = b_g_arr[bst_feat][0]
            best_features_beta[i][0] = bst_feat
            best_features_beta[i][1] = beta
            mask = arrList[i].T[bst_feat] > beta
            arrList[i * 2 + 1] = arrList[i][mask]
            arrList[i * 2 + 2] = arrList[i][~mask]
            print(i * 2 + 1, "\t", bst_feat, "\t", arrList[i * 2 + 1].shape[0])
            print(i * 2 + 2, "\t", bst_feat, "\t", arrList[i * 2 + 2].shape[0])
    else:
        print(i * 2 + 1, "\t", arrList[i * 2 + 1].shape[0], "\t", "--")
        print(i * 2 + 2, "\t", arrList[i * 2 + 2].shape[0], "\t", "--")
    if (i // (pow(2, iter + 2) - 2)):
        iter += 1
        print("-----------")
```
